## Nova consulta

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url2, sep =';')
dados
```

```python
dados['Valor_por_mes'] = dados['Valor']+ dados['Condominio']
```

```python
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] *12 + dados['IPTU']
```

```python
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quartos e ' + \
                     dados['Vagas'].astype(str) + ' vagas'
```

```python
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
```
