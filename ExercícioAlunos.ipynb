## exercícios parte 2

```python
dados= pd.read_csv('alunos.csv')
```

```python
dados = dados.fillna(dados['Notas'].mean())
```

```python
dados['Pontos_extra'] = dados['Notas']*0.4
```

```python
dados.head()
```

```python
dados['Nota_Final'] = (dados['Notas'] + dados['Pontos_extra']).clip(upper= 10)
```

```python
dados.head()
```

```python
dados['Aprovados_final'] = dados['Nota_Final'].apply(lambda x: True if x >= 6 else False)
```

```python
dados.head()
```

```python
selecao = (dados['Aprovado'] == False) & (dados['Aprovados_final']==True)
```
